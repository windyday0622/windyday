```python
def get_category(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    elif age <= 2:
        return 'Baby'
    elif age <= 12:
        return 'Child'
    elif age <= 17:
        return 'Teenager'
    elif age <= 35:
        return 'Young Adult'
    elif age <= 64:
        return 'Adult'
    else:
        return 'Elderly'

age = 25
category = get_category(age)
print(f"Age {age} falls into the category: {category}")
```

```python
tdf['age_cat'] = tdf.age.apply(lambda x:get_category(x))
tdf.head()
```
