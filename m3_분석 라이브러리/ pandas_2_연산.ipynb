```python
import pandas as pd

# Correcting the syntax to use dictionaries for key-value pairs
st1 = pd.Series({'국어': 100, '영어': 80, '수학': 90})
st2 = pd.Series({'수학': 80, '국어': 90, '영어': 80})

print(st1,'\n')
print(st2)
```

```python
pd.DataFrame(st1+st2)
```

```python
# 두 학생의 과목별 점수로 사칙연산 수행
add = st1 + st2
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2)
df = pd.concat([add,sub,mul,div],axis=1)
print(df)
```

```python
df = pd.concat([add,sub,mul,div], axis=1)
df.columns = ['덧셈', '뺄셈', '곱셈', '나눗셈']
df
```

```python
import pandas as pd
import numpy as np  # Import NumPy

# Define two pandas Series objects
st1 = pd.Series({'국어': np.nan, '영어': 80, '수학': 90})
st2 = pd.Series({'수학': 80, '국어': 90})

# Perform arithmetic operations
add = st1 + st2  # Addition
sub = st1 - st2  # Subtraction
mul = st1 * st2  # Multiplication
div = round((st1 / st2), 2)  # Division

# Store the results in a DataFrame
result = pd.DataFrame([add, sub, mul, div],
                      index=['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

#### 그룹 연산
* 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함
* 그룹 연산은 데이터를 집계, 변환, 필터링하는데 효율적이며 판다스 groupby() 메소드를 사용함
* 그룹 객체 만들기(분할)
* 그룹 연산 메소드(적용-결합)

```python
import pandas as pd
import seaborn as sns

# titanic 데이터 셋에서 age, sex 등 열을 선택하여 데이터 프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

print('승객 수:', len(df))
print(df.head())
```
