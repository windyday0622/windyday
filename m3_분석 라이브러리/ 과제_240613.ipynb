Q. 다음 사항을 수행하세요.
- 평균 50, 표준편차 10인 정규분포에서 100개 샘플 생성
- 기본 데이터의 평균과 중앙값 계산
- 데이터에 이상치 500 추가
- 이상치가 포함된 데이터의 평균과 중앙값 계산
- 이상치 포함에 따른 평균값과 중앙값간의 변화 설명
- pandas를 이용해서 위의 과제를 수행하세요.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 평균 50, 표준편차 10인 정규분포에서 100개 샘플 생성
np.random.seed(0)
data = np.random.normal(50, 10, 100)

# 기본 데이터의 평균과 중앙값 계산
mean_original = np.mean(data)
median_original = np.median(data)

# 데이터에 이상치 500 추가
data_with_outlier = np.append(data, 500)

# 이상치가 포함된 데이터의 평균과 중앙값 계산
mean_with_outlier = np.mean(data_with_outlier)
median_with_outlier = np.median(data_with_outlier)

# 결과 출력
print("기본 데이터의 평균:", mean_original)
print("기본 데이터의 중앙값:", median_original)
print("이상치 포함 데이터의 평균:", mean_with_outlier)
print("이상치 포함 데이터의 중앙값:", median_with_outlier)

# 이상치 포함에 따른 평균값과 중앙값의 변화 설명
mean_change = mean_with_outlier - mean_original
median_change = median_with_outlier - median_original

print("이상치 포함에 따른 평균값 변화:", mean_change)
print("이상치 포함에 따른 중앙값 변화:", median_change)

# 데이터 시각화
plt.figure(figsize=(14, 7))

# 기본 데이터의 히스토그램
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color='blue', alpha=0.7, label='Original Data')
plt.axvline(mean_original, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean_original:.2f}')
plt.axvline(median_original, color='g', linestyle='dashed', linewidth=1, label=f'Median: {median_original:.2f}')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# 이상치 포함 데이터의 히스토그램
plt.subplot(1, 2, 2)
plt.hist(data_with_outlier, bins=20, color='orange', alpha=0.7, label='Data with Outlier')
plt.axvline(mean_with_outlier, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean_with_outlier:.2f}')
plt.axvline(median_with_outlier, color='g', linestyle='dashed', linewidth=1, label=f'Median: {median_with_outlier:.2f}')
plt.title('Data Distribution with Outlier')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
