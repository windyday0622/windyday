```python
# Q.간단한 데이터프레임 작성 후 csv, json, pickle 형태로 저장 후 다시 불러와서 출력하세요.

import pandas as pd

# 데이터프레임 생성
data = {'이름': ['철수', '영희', '민수'],
        '나이': [30, 25, 35],
        '성별': ['남', '여', '남']}

df = pd.DataFrame(data)

# CSV로 저장 후 불러오기
df.to_csv('data.csv', index=False)
df_csv = pd.read_csv('data.csv')
print("CSV 파일로 저장 후 불러오기:")
print(df_csv)

# JSON으로 저장 후 불러오기
df.to_json('data.json', orient='records')
df_json = pd.read_json('data.json')
print("\nJSON 파일로 저장 후 불러오기:")
print(df_json)

# Pickle로 저장 후 불러오기
df.to_pickle('data.pkl')
df_pickle = pd.read_pickle('data.pkl')
print("\nPickle 파일로 저장 후 불러오기:")
print(df_pickle)
```
