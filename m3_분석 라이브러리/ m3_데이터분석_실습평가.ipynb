Q2. data_analysis.csv를 기반으로 아래 예시를 참조하여 새로운 분석 과제를 정의한 후, 그 과제에 적합한 분석용 데이터셋을 만든 후, 과제에 적합한 이유 및 특징을 기술하고 data_obj.csv로 저장하세요.

(예시)고객별 구매감소 원인을 파악하기 위한 분석형 데이터셋을 생성한 후 탐색적 분석을 수행.(거래기준 데이터 -> 고객기준 데이터로 변환)

```python
import pandas as pd
import matplotlib.pyplot as plt

# 개별 파일 경로 지정
file_path1 = '/content/data_rev1.csv'
file_path2 = '/content/data_rev2.csv'
file_path3 = '/content/data_rev3.csv'

# 데이터 로드
data_rev1 = pd.read_csv(file_path1)
data_rev2 = pd.read_csv(file_path2)
data_rev3 = pd.read_csv(file_path3)

# 데이터 확인
print("data_rev1 데이터 구조:")
print(data_rev1.head())
print("\ndata_rev2 데이터 구조:")
print(data_rev2.head())
print("\ndata_rev3 데이터 구조:")
print(data_rev3.head())
```

```python
print("\ndata_rev1 컬럼:")
print(data_rev1.columns)

print("\ndata_rev2 컬럼:")
print(data_rev2.columns)

print("\ndata_rev3 컬럼:")
print(data_rev3.columns)
```

```python
# 날짜 형식 변환 및 'year_month' 열 생성
data_rev3['date'] = pd.to_datetime(data_rev3['date'])
data_rev3['year_month'] = data_rev3['date'].dt.to_period('M')

# data_rev1과 data_rev3 병합
data_rev = pd.merge(data_rev1, data_rev3[['no', 'year_month']], on='no')

# 월별 제품 판매량 집계
monthly_sales = data_rev.groupby(['year_month', 'product_20', 'product_21'])[['quantity_20', 'quantity_21']].sum().unstack()

# 시각화
monthly_sales.plot(kind='line', figsize=(14, 7), title='Monthly product sales trend')
plt.xlabel('Monthly period')
plt.ylabel('Sales Volume')
plt.legend(title='Product')
plt.show()

print("인사이트 1: 월별 제품 판매 추이가 반영되었습니다.")
```

```python
# 날짜별 총 판매량 집계
data_rev['total_quantity_20'] = data_rev['quantity_20']
data_rev['total_quantity_21'] = data_rev['quantity_21']
daily_sales = data_rev.groupby('date')[['total_quantity_20', 'total_quantity_21']].sum()

# 판매량이 가장 많은 날짜 상위 5개
peak_dates_20 = daily_sales.nlargest(5, 'total_quantity_20')
peak_dates_21 = daily_sales.nlargest(5, 'total_quantity_21')

# 시각화
daily_sales['total_quantity_20'].plot(kind='line', figsize=(14, 7), title='Daily gross sales volume trend (2020)')
plt.scatter(peak_dates_20.index, peak_dates_20['total_quantity_20'], color='red')
plt.xlabel('Dates')
plt.ylabel('Total sales')
plt.axvline(x=peak_dates_20.index.max(), color='red', linestyle='--')
plt.show()

daily_sales['total_quantity_21'].plot(kind='line', figsize=(14, 7), title='Daily gross sales volume trend (2021)')
plt.scatter(peak_dates_21.index, peak_dates_21['total_quantity_21'], color='red')
plt.xlabel('Dates')
plt.ylabel('Total sales')
plt.axvline(x=peak_dates_21.index.max(), color='red', linestyle='--')
plt.show()

print("인사이트 2: 특정 날짜의 판매 피크가 반영되었습니다.")
```

```python
# 평균 가격 계산
data_rev['average_price_20'] = data_rev['price_20']
data_rev['average_price_21'] = data_rev['price_21']

# 월별 평균 가격 집계
average_price = data_rev.groupby(['year_month', 'product_20', 'product_21'])[['average_price_20', 'average_price_21']].mean().unstack()

# 시각화
average_price.plot(kind='line', figsize=(14, 7), title='Monthly product average price trend')
plt.xlabel('A monthly period')
plt.ylabel('Average Price')
plt.legend(title='Product')
plt.show()

# 가격과 판매량의 상관관계 분석
correlation_20 = data_rev[['average_price_20', 'quantity_20']].corr()
correlation_21 = data_rev[['average_price_21', 'quantity_21']].corr()
print("2020년 가격과 판매량의 상관관계:")
print(correlation_20)
print("2021년 가격과 판매량의 상관관계:")
print(correlation_21)

print("인사이트 3: 제품별 가격 변동과 판매량 상관관계가 반영되었습니다.")
```

```python
# 인사이트가 반영된 DataFrame 저장
data_rev.to_csv('data_analysis.csv', index=False)
print("data_analysis.csv 파일이 저장되었습니다.")
```

##새로운 분석 과제에 대한 데이터셋 설명

# 과제에 적합한 이유
- 목적성:월별 평균 판매 가격을 파악하여 가격 변동의 추세를 분석하는 것은 중요한 비즈니스 인사이트를 제공함.
- 시간 기반 분석: 월별 데이터를 사용하여 장기적인 가격 변동 패턴을 식별할 수 있음.

# 데이터셋의 특징
- 연속성: 각 월별 데이터가 있어 시간의 흐름에 따른 변화를 쉽게 관찰할 수 있음.
- 평균 가격: 월별 평균 판매 가격을 계산함으로써 월별 가격 변동의 전반적인 추세를 파악할 수 있음.

#데이터 구성
- year_month: 각 데이터 항목의 년월
- average_price_20: 해당 월의 product_20의 평균 판매 가격

```python
import pandas as pd

file_path = '/content/data_analysis.csv'

data = pd.read_csv(file_path)

# 필요한 열 선택
data_monthly = data[['year_month', 'average_price_20']]

# 'year_month' 기준으로 그룹화하고 'average_price_20'의 평균을 계산
monthly_average_price = data_monthly.groupby('year_month').mean().reset_index()

# 결과 데이터셋을 CSV 파일로 저장
output_file_path = 'data_obj.csv'
monthly_average_price.to_csv(output_file_path, index=False)

print(f"Data saved to {output_file_path}")

# 결과 데이터셋 확인
print(monthly_average_price.head())
```

```python
# 그래프 시각화
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/data_obj.csv'

data = pd.read_csv(file_path)

# 필요한 열 선택
data_monthly = data[['year_month', 'average_price_20']]

# 'year_month' 기준으로 그룹화하고 'average_price_20'의 평균을 계산
monthly_average_price = data_monthly.groupby('year_month').mean().reset_index()

# 시각화를 위한 데이터 준비
monthly_average_price['year_month'] = pd.to_datetime(monthly_average_price['year_month'])

# 시각화 설정
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_average_price, x='year_month', y='average_price_20', marker='o')

plt.title('Monthly Average Price of Product 20')
plt.xlabel('Year-Month')
plt.ylabel('Average Price 20')
plt.xticks(rotation=45)

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
# 히트맵
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 'year_month'를 datetime 형식으로 변환
monthly_average_price['year_month'] = pd.to_datetime(monthly_average_price['year_month'])

# 연도와 월을 따로 추출
monthly_average_price['year'] = monthly_average_price['year_month'].dt.year
monthly_average_price['month'] = monthly_average_price['year_month'].dt.month

# 피벗 테이블 생성
pivot_table = monthly_average_price.pivot(index='month', columns='year', values='average_price_20')

# 시각화 설정
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap='coolwarm')

# 제목과 라벨 설정
plt.title('Heatmap of Monthly Average Price of Product 20')
plt.xlabel('Year')
plt.ylabel('Month')

# 그래프 표시
plt.tight_layout()
plt.show()
```
