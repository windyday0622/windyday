Task1_0617. 선그래프, 히스토그램, 산점도, 막대그래프 4개의 subplot으로 구성된 2 x 2 그래프를 작성하세요.(데이터는 알아서 생성하기, matplotlib를 사용해서)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.normal(size=1000)
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# 선그래프
axs[0, 0].plot(x, y, label='Sine Wave')
axs[0, 0].set_title('Line Ghaph')
axs[0, 0].legend()

# 히스토그램
axs[0, 1].hist(data, bins=30, alpha=0.75, color='green')
axs[0, 1].set_title('Histogram')

# 산점도
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
axs[1, 0].set_title('Scatterplot')

# 막대그래프
axs[1, 1].bar(categories, values, color='blue')
axs[1, 1].set_title('Bar graph')

plt.tight_layout()
plt.show()
```
