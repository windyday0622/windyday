Task4. 2023년 1월 1일부터 2023년 12월 31일까지의 날짜 범위를 생성하고, 이 날짜 범위에서 매월 첫 번째 날만 포함하는 데이터프레임을 작성하세요.

```python
import pandas as pd

# 날짜 범위 정의
start_date = '2023-01-01'
end_date = '2023-12-31'

# 날짜 범위 생성 (매월 첫 번째 날)
date_range = pd.date_range(start=start_date, end=end_date, freq='MS')

# 데이터프레임 생성
df = pd.DataFrame(date_range, columns=['Date'])

print(df)
```

Task5. 2020년부터 2025년까지의 분기별 기간 범위를 생성하고, 각 분기의 시작과 끝 날짜를 포함하는 데이터프레임을 작성하세요.

```python
import pandas as pd
from pandas.tseries.offsets import QuarterEnd

# 시작 년도와 끝 년도 정의
start_year = 2020
end_year = 2025

# 분기 정보를 저장할 리스트 생성
quarters = []

# 년도와 분기를 순회하며 시작 날짜와 끝 날짜 계산
for year in range(start_year, end_year + 1):
    for quarter in range(1, 5):
        start_date = pd.Timestamp(year, (quarter - 1) * 3 + 1, 1)
        end_date = (start_date + QuarterEnd(n=1)).normalize()
        quarters.append([f"Q{quarter}-{year}", start_date, end_date])

# 데이터프레임 생성
quarters_df = pd.DataFrame(quarters, columns=["분기", "시작 날짜", "끝 날짜"])

# 데이터프레임 출력
quarters_df
```

Task6. 0일부터 10일까지의 시간 범위를 12시간 간격으로 생성하고, 각 간격에 대해 누적 시간을 계산하여 데이터프레임을 작성하세요.

```python
import pandas as pd

# Define the time range and interval
time_range = range(0, 11)  # 0 to 10
interval = 12  # 12-hour interval

# Create a list to store the cumulative times
cumulative_times = []

# Calculate cumulative times at each interval
for i in time_range:
    cumulative_times.append(i * interval)

# Create a DataFrame
df = pd.DataFrame({
    'Time (days)': time_range,
    'Cumulative Time (hours)': cumulative_times
})

df
```

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

# 분기 계산을 위한 함수 정의
def get_quarter(month):
    if month in [1, 2, 3]:
        return 'Q1'
    elif month in [4, 5, 6]:
        return 'Q2'
    elif month in [7, 8, 9]:
        return 'Q3'
    else:
        return 'Q4'

# 'quarter' 열 추가
df['quarter'] = df['date'].dt.month.apply(get_quarter)

# 분기별 데이터의 평균 계산
quarterly_avg = df.groupby('quarter')['value'].mean()

print(quarterly_avg)
```
